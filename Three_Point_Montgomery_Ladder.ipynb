```python
#Three point montgomery ladder P+kQ, proof of concept
def three_point_ladder():
    p=2^372*3^239-1 
    E=EllipticCurve(GF(p),[1,0])
    p_xcoordinate=int(raw_input("What is the x-coordinate of the first point?\n"))
    if(kronecker_symbol(p_xcoordinate^3+p_xcoordinate,p)!=1):
        print("Such a point doesn't exist on the curve\n")
        return -1
    q_xcoordinate=int(raw_input("What is the x-coordinate of the second point?\n"))
    if(kronecker_symbol(q_xcoordinate^3+q_xcoordinate,p)!=1):
        print("Such a point doesn't exist on the curve\n")
        return -2
    P=E(p_xcoordinate,Mod(p_xcoordinate^3+p_xcoordinate,p).sqrt())
    Q=E(q_xcoordinate,Mod(q_xcoordinate^3+q_xcoordinate,p).sqrt())
    P_Q=int((P-Q)[0])
    user=raw_input("what is the random integer you have chosen?\n")
    c=int(user)
    if(c==0):
        print(P)
    else:
        digit_store=[]
        while(c):
            digit_store.append(c&1)
            c=c//2
        digit_store.reverse()
        flag=0
        A_X=0
        A_Z=0
        B_X=int(Q[0])
        B_Z=1
        C_X=int(P[0])
        C_Z=1
        while (len(digit_store)!=0):
            c=digit_store[0]
            temp_A_X=A_X
            temp_A_Z=A_Z
            temp_B_X=B_X
            temp_B_Z=B_Z
            temp_C_X=C_X
            temp_C_Z=C_Z
            if(c&1==1 and flag==0):
                A_X=int(Q[0])
                A_Z=1
                # DOUBLE B
                B_X,B_Z=(temp_B_X^2-temp_B_Z^2)^2,4*temp_B_X*temp_B_Z*(temp_B_X^2+temp_B_Z^2)
                # ladder on C=dadd(B,C,Q-P)
                C_X,C_Z=(temp_B_X*temp_C_X-temp_B_Z*temp_C_Z)^2,P_Q*(temp_B_X*temp_C_Z-temp_B_Z*temp_C_X)^2
                flag=1
            elif(c&1==0 and flag==1):
                # DOUBLE A
                A_X,A_Z=(temp_A_X^2-temp_A_Z^2)^2,4*temp_A_X*temp_A_Z*(temp_A_X^2+temp_A_Z^2)
                # ladder on B=dadd(A,B,Q)
                B_X,B_Z=(temp_A_X*temp_B_X-temp_A_Z*temp_B_Z)^2,2*(temp_A_X*temp_B_Z-temp_A_Z*temp_B_X)^2
                # ladder on C=dadd(A,C,P)
                C_X,C_Z=(temp_A_X*temp_C_X-temp_A_Z*temp_C_Z)^2,(temp_A_X*temp_C_Z-temp_A_Z*temp_C_X)^2
            elif(c&1==1 and flag==1):
                # ladder on A=dadd(A,B,Q)
                A_X,A_Z=(temp_A_X*temp_B_X-temp_A_Z*temp_B_Z)^2,2*(temp_A_X*temp_B_Z-temp_A_Z*temp_B_X)^2
                # DOUBLE B
                B_X,B_Z=(temp_B_X^2-temp_B_Z^2)^2,4*temp_B_X*temp_B_Z*(temp_B_X^2+temp_B_Z^2)
                # ladder on C=dadd(B,C,Q-P)
                C_X,C_Z=(temp_B_X*temp_C_X-temp_B_Z*temp_C_Z)^2,P_Q*((temp_B_X*temp_C_Z)-(temp_B_Z*temp_C_X))^2
            elif(c&1==0 and flag==0):
                continue
            digit_store.pop(0)
            A_X%=p
            A_Z%=p
            B_X%=p
            B_Z%=p
            C_X%=p
            C_Z%=p
    return((C_X*(C_Z.inverse_mod(p)))%p)
print(three_point_ladder())
```
